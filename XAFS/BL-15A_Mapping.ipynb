```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_scalebar.scalebar import ScaleBar
import os
os.chdir("C:\\Users\\cocoa\\Documents\\実験\\BL-15A\\220605_BL15A")
print(os.getcwd())
```

```python
data = pd.read_csv('Oharai1_copepod_LavaeM3_CH8.txt', header=None, index_col=None)
test=data.values
x=[]
for i in range (len(test)):
    j= max(test[i])
    x.append(j)
print(max(x))#カウントの最大値
```

### scalebarメモ
- dx: 1ピクセル当たりの長さに相当、(画像の幅の実際の長さ)/(dataの列の長さ)により算出

```python

fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)

sns.heatmap(data, cmap="jet",  vmax=max(x), vmin=0)
ax.invert_xaxis()
ax.invert_yaxis()
ax.set_yticks([])
ax.set_xticks([])
ax.text(0.05,0.8,"Fe",size=45,transform=ax.transAxes, color="white")

scalebar = ScaleBar(dx=0.01,
        units="mm",dimension="si-length",length_fraction=0.5,
        location="lower right",pad=0.5,border_pad=0.8,sep=None,color=None,box_color=None,scale_loc="top")
ax.add_artist(scalebar)
```
